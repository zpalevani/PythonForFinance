```python
from matplotlib import dates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf

```

```python
raw = yf.download('SPY AAPL', start='2010-01-01', end='2024-12-31')
```

```python
import matplotlib.pyplot as plt

# Plot SPY and AAPL closing prices
plt.figure(figsize=(12, 6))  # Set the figure size

raw['Close']['SPY'].plot(label='SPY', linewidth=2)
raw['Close']['AAPL'].plot(label='AAPL', linewidth=2)

# Title and labels
plt.title('Stock Prices of SPY & AAPL (2010-2019)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Closing Price (USD)', fontsize=12)
plt.legend()  # Show the legend
plt.grid(True)  # Show the grid

# Show the plot
plt.show()
```

**Key Observations:**

*   SPY (Blue Line)

 *   SPY represents the S&P 500 ETF, which tracks the overall stock market.
 *   It steadily increased from around 100 dollar in 2010 to over $300 by 2019.

 *   The upward trend indicates strong market growth over this period.

 *   Some dips are visible, likely due to market corrections or economic events.

*   AAPL (Orange Line)

 *   AAPL represents Apple Inc.'s stock price.
 *   It started at a much lower price than SPY under 50 dollar.
 *   Apple’s stock price grew significantly over the decade, crossing 250 dollar
  by 2019.
 *   The sharp increase from 2017-2019 suggests strong performance possibly from iPhone sales, services, or other product launches.

**What This Means:**

*   Both stocks trended upward  → Positive growth for both the market SPY and Apple AAPL.
*   AAPL had higher relative growth → Apple outperformed the general market over the decade.
*   Small dips/corrections in both → Likely due to economic downturns or company-specific factors.
